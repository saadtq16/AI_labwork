```
import pandas as pd
import numpy as np
from pandas import ExcelWriter
from pandas import ExcelFile
import math
from scipy.stats import mode

MAX = 100000

#reading training and test sets using pandas library
TrainSet = pd.read_excel("TrainingSet.xlsx")
TestSet = pd.read_excel("TestSet1.xlsx")
print('Training set:')
print(TrainSet)
#using for display purposes
col1 = test_set.iloc[:, 0].values
col2 = test_set.iloc[:, 1].values
col3 = test_set.iloc[:, 2].values
col4 = test_set.iloc[:, 3].values


def KNN(train_set, test_set, k):
  #separating features into respective numpy arrays
  leaf_len = train_set.iloc[:, 0].values
  leaf_wid = train_set.iloc[:, 1].values
  flower_len = train_set.iloc[:, 2].values
  flower_wid = train_set.iloc[:, 3].values
  label = train_set.iloc[:, 4].values

  #creating test set arrays
  ll = test_set.iloc[:, 0].values
  lw = test_set.iloc[:, 1].values
  fl = test_set.iloc[:, 2].values
  fw = test_set.iloc[:, 3].values

  #calculating euclidean distances of each test point
  dist = np.zeros([len(test_set), len(train_set)])
  for i in range(0,len(test_set)):
    for j in range(0,len(train_set)):
      eucl = math.sqrt((ll[i]-leaf_len[j])**2 + (lw[i]-leaf_wid[j])**2 + (fl[i]-flower_len[j])**2 + (fw[i]-flower_wid[j])**2)
      dist[i,j] = eucl

  #create array for storing labels based on neighbours
  buffer = np.empty([len(test_set), k], dtype='O')

  #finding nearest neighbours
  temp = dist
  for i in range(0,len(test_set)):
    for index in range(0,k):
      neighbour = np.min(temp[i])
      target = np.where(temp[i] == neighbour)
      #flush label value to buffer array
      buffer[i,index] = label[target[0][0]]
      #replace the current neighbour with MAX to look a the next neighbour
      temp[i][target[0][0]] = MAX
  #print(buffer)

  #perform working on buffer array
  working = np.zeros([len(test_set), k])
  for i in range(0,int(buffer.shape[0])):
    for j in range(0, int(buffer.shape[1])): 
      if buffer[i,j] == 'Arctica':
        working[i,j] = 1
      if buffer[i,j] == 'Harlequin':
        working[i,j] = 2
      if buffer[i,j] == 'Carolinian':
        working[i,j] = 3

  #find the majority label
  mod_arr = np.zeros([len(test_set)])
  for i in range(0,int(buffer.shape[0])):
    md = mode(working[i])
    mod_arr[i] = md[0][0]

  #reverse converting working array to decision label calculated from mode
  decision = np.empty([len(test_set)], dtype='O')
  for i in range(0,int(buffer.shape[0])):
    if mod_arr[i] == 1:
      decision[i] = 'Arctica'
    if mod_arr[i] == 2:
      decision[i] = 'Harlequin'
    if mod_arr[i] == 3:
      decision[i] = 'Carolinian'
  return decision

#run algorith with k=3
k3 = KNN(TrainSet, TestSet, 3)
#run algorith with k=5
k5 = KNN(TrainSet, TestSet, 5)
#run algorith with k=7
k7 = KNN(TrainSet, TestSet, 7)

#making a dataframe
print('\nResults of Test Set:')
df = pd.DataFrame(col1, columns = ['Leaf Length'])
df['Leaf Width'] = col2
df['Flower Length'] = col3
df['Flower Width'] = col4
df['Plant (k=3)'] = k3
df['Plant (k=5)'] = k5
df['Plant (k=7)'] = k7
print(df)

#flushing to an excel file
df.to_excel('KNN_results.xlsx')
```

```
import pandas as pd
import numpy as np
from pandas import ExcelWriter
from pandas import ExcelFile

train_set = pd.read_excel("TrainingSet.xlsx")
test_set = pd.read_excel("TestSet1.xlsx")
print(train_set)

#prepping data for training
X = train_set.iloc[:, 0:4].values
y = train_set.iloc[:, 4].values
X_test = test_set.iloc[:, 0:4].values

#feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X)
X_test = sc.fit_transform(X_test)

#training
from sklearn.ensemble import RandomForestClassifier

#default - 100 trees
classifier = RandomForestClassifier(n_estimators=100)
classifier.fit(X_train, y)
pred100 = classifier.predict(X_test)

#300 trees
classifier2 = RandomForestClassifier(n_estimators=300)
classifier2.fit(X_train, y)
pred300 = classifier.predict(X_test)

#500 trees
classifier3 = RandomForestClassifier(n_estimators=500)
classifier3.fit(X_train, y)
pred500 = classifier.predict(X_test)

#making a dataframe
df = pd.DataFrame(pred100, columns = ['Forest size = 100'])
df['Forest size = 300'] = pred300
df['Forest size = 500'] = pred500
print(df)

#flushing to an excel file
df.to_excel('results.xlsx')
```

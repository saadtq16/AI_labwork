```
import numpy as np
import math
from anytree import Node, RenderTree

def main():
  #Creating dataset according to question
  data_set = np.array([("rainy", "hot", "high", False, "no"), ("rainy", "hot", "high", True, "no"), ("overcast", "hot", "high", False, "yes"), ("sunny", "mild", "high", False, "yes"),
                      ("sunny", "cool", "normal", False, "yes"), ("sunny", "cool", "normal", True, "no"), ("overcast", "cool", "normal", True, "yes"), ("rainy", "mild", "high", False, "no"),
                      ("rainy", "cool", "normal", False, "yes"), ("sunny", "mild", "normal", False, "yes"), ("rainy", "mild", "normal", True, "yes"), ("overcast", "mild", "high", True, "yes"),
                      ("overcast", "hot", "normal", False, "yes"), ("sunny", "mild", "high", True, "no")])

  print('Data Set: ')
  #print(data_set)
  #asking user for data set information
  dim = int(input("Dimensions of data: "))
  target = int(input("Target index: "))
  classes = int(input("No. of classes: "))
  names = []
  for i in range(0, classes):
    names.append(input("Attribute title: "))
  print(names)
  entries = int(data_set.shape[0])

  #separating label attribute
  label = []
  for i in range(0,entries):
    label.append(data_set[i,target])
  label_set = set(label)
  print(label)

  #creating attributes
  attr = [[] for _ in range(dim-1)]
  for i in range(0, entries):
    for j in range(0, len(attr)):
      attr[j].append(data_set[i,j])
  print(attr)

  #making unique set of attributes
  sets = []
  for i in range(0, len(attr)):
    sets.append(set(attr[i]))
  print(sets)
  print(label_set)
  print(len(sets[0]), len(sets[2]))

  #calculating overall entropy
  label_count = []
  for x in label_set:
    count = 0
    for y in label:
      if y == x:
        count += 1
    label_count.append(count)
  print(label_count)
  total_entropy = calc_entropy(label_count)
  print(total_entropy)

  #calculating entropy for all classes of attributes
  attr_count = [[] for _ in range(len(attr))]
  for i in range(0, len(sets)):
    index = 0
    attr_count[i] = [[] for _ in range(len(sets[i]))]
    #print(sets[i])
    for j in sets[i]:
      #print(j)
      for k in label_set:
        count = 0
        #print(k)
        for l in range(0, entries):
          if attr[i][l] == j and label[l] == k:
            count += 1
            #print(count)
        attr_count[i][index].append(count)
      index += 1
  print(attr_count)

  #calculating entropies of all counts
  E = [[] for _ in range(len(attr))]
  for i in range(0, len(attr_count)):
    for j in range(0, len(attr_count[i])):
      e = calc_entropy(attr_count[i][j])
      E[i].append(e)
  print(E)
  
  #calculating Information Gain
  I = [[] for _ in range(len(attr))]
  for i in range(0, len(attr_count)):
    temp = 0
    for j in range(0, len(attr_count[i])):
      temp += (sum(attr_count[i][j]) / entries) * E[i][j]
    I[i].append(round(temp,3))
  print(I)

  #calculating gains
  G = [[] for _ in range(len(I))]
  for  i in range(0, len(I)):
    gain = round(total_entropy - I[i][0], 3)
    G[i].append(gain)
  print(G)

  #determining root node
  for i in range(0, len(G)):
    if G[i] == max(G):
      highest = i
  root = Node(names[i]+"") 

def calc_entropy(labels):
  buff = []
  tot = 0
  for i in labels:
    tot += i
  for j in range(0, len(labels)):
    buff.append(labels[j] / tot)  
  entropy = 0
  for k in buff:
    if k == 0:
      return 0
    entropy += (-(k) * math.log(k,2))
  return round(entropy, 3)

main()
```

```
#setting initial values of alpha and beta
A = 100000
B = -100000

def prune(d, node, tree, maxPlayer, alpha, beta, max_height):
  if d == max_height:
    return tree[node]
  
  if maxPlayer:                       #MAX player's turn
    #make current best as infinite min
    best = B
    #loop to traverse through all children of the current node
    for i in range(0, max_height-1):
      val = prune(d+1, (node*2)+i, tree, False, alpha, beta, max_height)
      best = max(best,val)              #choose best max amongst children
      MAX = max(alpha, best)            #MAX will take value of best since alpha is infinite minimum
      
      #shortening traversal by applying alpha-beta pruning rule 
      if alpha >= beta:
        break
    return best
  
  else:                               #MIN player's turn
    #make current best as infinite max
    best = A
    #loop to traverse through all children of the current node
    for i in range(0, max_height-1):
      val = prune(d+1, (node*2)+i, tree, True, alpha, beta, max_height)
      best = min(best,val)             #choose best min amongst children
      MIN = min(beta,best)             #MIN will take value of best since beta is infinite maximum
      
      #shortening traversal by applying alpha-beta pruning rule 
      if alpha >= beta:
        break
    return best

#main program
tree = [7, -5, 6, -9, 1, 2, 0, 8]       #feed a binary tree
tree_height = 3
print("The optimal value is :", prune(0, 0, tree, True, B, A, tree_height))  
      
```

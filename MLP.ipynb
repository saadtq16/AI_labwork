```
############# MLP Implementation for XOR operator (from scratch) #############
import pandas as pd
#defining Theta
theta = 1

#nuron function which accepts the weights and inputs in an list
def neuron(W,X):
  dec = 0
  res = 0
  for i in range(0,len(W)):
    prod = W[i] * X[i]
    res += prod
  #checking unit-step condition
  if res >= theta:
    dec = 1
  else:
    dec = 0
  return dec

def NeuralNetwork(x,y):
  #builidng a three neuron model
  Inputs = [x,y]
  #defining weights
  W_A = [-1,1]
  W_B = [1,-1]
  W_O = [1,1]
  A = neuron(W_A,Inputs)
  B = neuron(W_B,Inputs)
  Mid = [A,B]
  O = neuron(W_O,Mid)
  return O

#running neural network on all possible 2 input combinations

x = [0,0,1,1]
y = [0,1,0,1]
ans = []
for i in range(0,len(x)):
  ans.append(NeuralNetwork(x[i],y[i]))
print('XOR Table:\n')
df = pd.DataFrame(x, columns = ['X'])
df['Y'] = y
df['Result'] = ans

print(df)

```

```
import pandas as pd
import numpy as np
from pandas import ExcelWriter
from pandas import ExcelFile
from sklearn.neural_network import MLPClassifier

#reading training and test sets using pandas library
train_set = pd.read_excel("TrainingSet.xlsx")
test_set = pd.read_excel("TestSet1.xlsx")
print('Training set:')
print(train_set)
#using for display purposes
col1 = test_set.iloc[:, 0].values
col2 = test_set.iloc[:, 1].values
col3 = test_set.iloc[:, 2].values
col4 = test_set.iloc[:, 3].values

#prepping data for training
X = train_set.iloc[:, 0:4].values
y = train_set.iloc[:, 4].values
X_test = test_set.iloc[:, 0:4].values

clf = MLPClassifier(solver = 'lbfgs', alpha=1e-5, hidden_layer_sizes=(7,7), random_state=1).fit(X, y)
pred = clf.predict(X_test)

#making a dataframe
print('\nResults of Test Set using MLP classificiation:')
df = pd.DataFrame(col1, columns = ['Leaf Length'])
df['Leaf Width'] = col2
df['Flower Length'] = col3
df['Flower Width'] = col4
df['Plant'] = pred
print(df)

#flushing to an excel file
df.to_excel('MLP_results.xlsx')
```

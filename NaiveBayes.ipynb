```
import numpy as np

#Creating dataset according to question
data_set = np.array([("sales", "senior", "31-35", "46-50K"), ("sales", "junior", "26-30", "26-30K"), ("sales", "junior", "31-35", "31-35K"), ("systems", "junior", "21-25", "46-50K"),
                     ("systems", "senior", "31-35", "66-70K"), ("systems", "junior", "26-30", "46-50K"), ("systems", "senior", "41-45", "66-70K"), ("marketing", "senior", "36-40", "46-50K"),
                     ("marketing", "junior", "31-35", "41-45K"), ("secretary", "senior", "46-50", "36-40K"), ("secretary", "junior", "26-30", "26-30K")])

print('Data Set: ')
print(data_set)

#creating separate attributes
department = []
status = []
age = []
salary = []

#separating attributes
senior_count = 0
junior_count = 0

for i in range(0,11):
  dep = data_set[i,0]
  stat = data_set[i,1]
  a = data_set[i,2]
  sal = data_set[i,3]
  department.append(dep)
  status.append(stat)
  age.append(a)
  salary.append(sal)
  if data_set[i,1] == 'senior':
    senior_count += 1
  elif data_set[i,1] == 'junior':
    junior_count += 1

#probability of label
junior_prob = junior_count / (junior_count + senior_count)
senior_prob = senior_count / (junior_count + senior_count) 

#count unique classes
unique_dep = set(department)
unique_stat = set(status)
unique_age = set(age)
unique_sal = set(salary)
print('\nClasses: ')
print(unique_dep, unique_age, unique_sal)
print('\nLabel: ')
print(unique_stat)

#constants for Laplace Smoothing
a = 1
d = len(unique_dep)+len(unique_age)+len(unique_sal)

# frequency table for departments :
index_v = 0
dep_freq = np.empty((len(unique_dep), len(unique_stat)+1), dtype='O')
for x in unique_dep:
  dep_freq[index_v,0] = x
  index_h = 1
  for y in unique_stat:
    count = 0
    for i in range(0,11):
      if data_set[i,0] == x and data_set[i,1] == y:
        count += 1
    if y == 'junior':
      dep_freq[index_v,index_h+1] = (count + a) / (junior_count + a*d)
    elif y == 'senior':
      dep_freq[index_v,index_h] = (count + a) / (senior_count + a*d)
    #index_h += 1
  index_v += 1
print('\nConditional probabilities for department: ')
print(dep_freq)

# frequency table for ages :
index_v = 0
age_freq = np.empty((len(unique_age), len(unique_stat)+1), dtype='O')
for x in unique_age:
  age_freq[index_v,0] = x
  index_h = 1
  for y in unique_stat:
    count = 0
    for i in range(0,11):
      if data_set[i,2] == x and data_set[i,1] == y:
        count += 1
    if y == 'junior':
      age_freq[index_v,index_h+1] = (count + a) / (junior_count + a*d)
    elif y == 'senior':
      age_freq[index_v,index_h] = (count + a) / (senior_count + a*d)
    #index_h += 1
  index_v += 1
print('\nConditional probabilities for ages: ')
print(age_freq)

# frequency table for salary :
index_v = 0
sal_freq = np.empty((len(unique_sal), len(unique_stat)+1), dtype='O')
for x in unique_sal:
  sal_freq[index_v,0] = x
  index_h = 1
  for y in unique_stat:
    count = 0
    for i in range(0,11):
      if data_set[i,3] == x and data_set[i,1] == y:
        count += 1
    if y == 'junior':
      sal_freq[index_v,index_h+1] = (count + a) / (junior_count + a*d)
    elif y == 'senior':
      sal_freq[index_v,index_h] = (count + a) / (senior_count + a*d)
    #index_h += 1
  index_v += 1
print('\nConditional probabilities for salary: ')
print(sal_freq)


#Testing for set 1
test_set = np.array([('marketing', '31-35', '46-50K')])
print('\nTest set entered: ')
print(test_set)

dep_test = test_set[0,0]
age_test  = test_set[0,1]
sal_test = test_set[0,2]

#getting respective probabilites
for i in range(len(dep_freq)):
  if dep_test == dep_freq[i,0]:
    dep_s = dep_freq[i,1]
    dep_j = dep_freq[i,2]

#print(dep_s,dep_j)

for i in range(len(age_freq)):
  if age_test == age_freq[i,0]:
    age_s = age_freq[i,1]
    age_j = age_freq[i,2]

#print(age_s,age_j)

for i in range(len(sal_freq)):
  if sal_test == sal_freq[i,0]:
    sal_s = sal_freq[i,1]
    sal_j = sal_freq[i,2]

#print(sal_s,sal_j)

prob_s = dep_s * age_s * sal_s * senior_prob
prob_j = dep_j * age_j * sal_j * junior_prob

print('\nNaive-Baye\'s classifier: ')
#verdict
if prob_s > prob_j:
  print('Status is senior')
elif prob_s < prob_j:
  print('Status is junior')
elif prob_s == prob_j:
  print('Status can not be determined')


#print('\nPercentage accuracy: ')
#acc = (abs(prob_s - prob_j) / prob_j) * 100 
#print(round(acc,2))
```

```
################## One vs All Classification ##################
import pandas as pd
import numpy as np
from pandas import ExcelWriter
from pandas import ExcelFile
from sklearn.multiclass import OneVsRestClassifier
from sklearn.svm import SVC

#reading training and test sets using pandas library
train_set = pd.read_excel("TrainingSet.xlsx")
test_set = pd.read_excel("TestSet1.xlsx")
print('Training set:')
print(train_set)
#using for display purposes
col1 = test_set.iloc[:, 0].values
col2 = test_set.iloc[:, 1].values
col3 = test_set.iloc[:, 2].values
col4 = test_set.iloc[:, 3].values

#prepping data for training
X = train_set.iloc[:, 0:4].values
y = train_set.iloc[:, 4].values
X_test = test_set.iloc[:, 0:4].values

#define model 
model = SVC()
OvR = OneVsRestClassifier(model)

#fitting model for training
OvR.fit(X,y)

#testing
pred = OvR.predict(X_test)

#making a dataframe
print('\nResults of Test Set using OnevAll classificiation:')
df = pd.DataFrame(col1, columns = ['Leaf Length'])
df['Leaf Width'] = col2
df['Flower Length'] = col3
df['Flower Width'] = col4
df['Plant'] = pred
print(df)
```

```
################## One vs One Classification ##################
import pandas as pd
import numpy as np
from sklearn.svm import SVC
from sklearn.multiclass import OneVsOneClassifier

#define model 
model = SVC()
OvO = OneVsOneClassifier(model)

#fitting model for training
OvO.fit(X,y)

#testing
pred2 = OvO.predict(X_test)

#making a dataframe
print('\nResults of Test Set using OnevOne classificiation:')
df = pd.DataFrame(col1, columns = ['Leaf Length'])
df['Leaf Width'] = col2
df['Flower Length'] = col3
df['Flower Width'] = col4
df['Plant'] = pred2
print(df)
```
